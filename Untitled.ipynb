```python
%matplotlib inline

import geopandas as gpd
import pandas as pd
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.io import shapereader
import matplotlib.pyplot as plt
import matplotlib
import geoplot as gplt
```

```python
#variables for reading in Natural Earth Data
resolution = '50m'
category = 'cultural'
name = 'admin_0_countries'

earth_boundary = shapereader.natural_earth(resolution, category, name)

# read the shapefile using geopandas
eb_df = gpd.read_file(earth_boundary)

# read australian borders
aus_boundary = eb_df.loc[eb_df['ADMIN'] == 'Australia']['geometry'].values[0]
```

```python
resolution = '50m'
category = 'cultural'
name = 'admin_1_states_provinces'

state_boundaries = shapereader.natural_earth(resolution, category, name)

# read the shapefile using geopandas
sb_df = gpd.read_file(state_boundaries)

# read australian borders
aus_sb = sb_df.loc[sb_df['iso_a2'] == 'AU']['geometry'].values
```

```python
#import populated places

resolution = '50m'
category = 'cultural'
name = 'populated_places'

places = shapereader.natural_earth(resolution, category, name)

# read the shapefile using geopandas
pd.set_option('display.max_columns', 500)
places_df = gpd.read_file(places)

# read australian populated places
aus_places = places_df.loc[places_df['SOV0NAME'] == 'Australia']#['geometry'].values
```

```python
#display

fig = plt.figure(facecolor='#c9d5d8', figsize=[15, 15])
ax = plt.axes(projection=ccrs.PlateCarree())
ax.background_patch.set_fill(False)
ax.set_extent([108, 160,-47, -8], crs=ccrs.PlateCarree())

ax.add_geometries(aus_sb, crs=ccrs.PlateCarree(), facecolor='#f4eecb', edgecolor='0.8')
ax.add_geometries(aus_boundary, crs=ccrs.PlateCarree(), facecolor='none', edgecolor='0.2')

aus_places.plot(ax=ax, marker='o', color='red', markersize=5, zorder=10)
```

```python
prisons = gpd.read_file("data/immi-detention-centres.geojson")

prisons = prisons.reindex([0,1,2,4,5,6,3])
prisons = prisons.reset_index()
prison_pop = prisons['total']
scale = prison_pop * 18
prisons
```

```python
###### display
matplotlib.rc('font', family='TakaoPGothic')

fig = plt.figure(facecolor='#c9d5d8', figsize=[25, 25])

ax = plt.axes(projection=ccrs.PlateCarree())

ax.background_patch.set_fill(False)
ax.set_extent([108, 168,-47, -8], crs=ccrs.PlateCarree())

ax.add_geometries(aus_sb, crs=ccrs.PlateCarree(), facecolor='#f4eecb', edgecolor='0.8')
ax.add_geometries(aus_boundary, crs=ccrs.PlateCarree(), facecolor='none', edgecolor='0.2')

ax.scatter(x=prisons.geometry.x, y=prisons.geometry.y, s=scale, c=prisons.children, cmap='Reds', vmin=-1.5, vmax=1.8, edgecolor='black', alpha=0.7, zorder=9)

#Create Legend
for pop in [10, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=pop*18, label=str(pop))
leg = ax.legend(scatterpoints=1, frameon=False, labelspacing=8, handlelength=14, handletextpad=0, borderpad=3, title_fontsize=13,
                title='Immigration Detention Centre Population \n ')

leg._legend_box.sep = 20
ax.set_title('Number of People in Onshore Immigration Detention Centres in Australia \n ', fontdict={'fontsize':18}, pad=15)

ax.annotate('', xy=(945,598), xycoords='axes points', xytext=(0,30), textcoords='offset points', arrowprops=dict(arrowstyle='->'), zorder=10)

#annotations
ax.annotate('MITA \n 242 people/242 \n Contains Children \n ', xy=(prisons.iloc[0].geometry.x, prisons.iloc[0].geometry.y), 
            xytext=(70, -45), textcoords='offset points', arrowprops=dict(arrowstyle='-'), zorder=10)
ax.annotate('Villawood Detention Centre \n 489 people/489', xy=(prisons.iloc[3].geometry.x, prisons.iloc[3].geometry.y), 
            xytext=(50, 0), textcoords='offset points', zorder=10)
ax.annotate('BITA \n 185 people/185', xy=(prisons.iloc[1].geometry.x, prisons.iloc[1].geometry.y), 
            xytext=(30, 0), textcoords='offset points', zorder=10)
ax.annotate('AITA \n 13 people/13', xy=(prisons.iloc[2].geometry.x, prisons.iloc[2].geometry.y), 
            xytext=(-60, -50), textcoords='offset points', arrowprops=dict(arrowstyle='-'), zorder=10)
ax.annotate('Northern APOD \n 15 people/15 \n Contains Children\n ', xy=(prisons.iloc[5].geometry.x, prisons.iloc[5].geometry.y), 
            xytext=(-130, 0), textcoords='offset points', arrowprops=dict(arrowstyle='-'), zorder=10)
ax.annotate('Yongah Hill Detention Centre \n 298 people/298', xy=(prisons.iloc[4].geometry.x, prisons.iloc[4].geometry.y), 
            xytext=(40, 0), textcoords='offset points', zorder=10)
ax.annotate('Perth IDC \n 298 people/298', xy=(prisons.iloc[6].geometry.x, prisons.iloc[6].geometry.y), 
            xytext=(-130, 0), textcoords='offset points', arrowprops=dict(arrowstyle='-'), zorder=10)

#credits
ax.annotate('Data from Australian Department of Home Affairs 28 Feb 2019 \n https://www.homeaffairs.gov.au/research-and-stats/files/immigration-detention-statistics-feb-2019.pdf', xy=(3,3), xycoords='axes points')
```

```python
fig.savefig('aus-immi-detention-map.png', facecolor='#c9d5d8', bbox_inches='tight')
```
